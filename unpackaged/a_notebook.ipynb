### Seaborn experiment ###

```python
sharing_config = False

import warnings
if sharing_config:
    warnings.filterwarnings('ignore')
else:
    warnings.resetwarnings()

import seaborn as sns
import pandas as pd
import numpy as np

# Create some sample data
data = pd.DataFrame({'time': pd.date_range('2022-01-01', periods=100), 
                     'value': np.random.normal(size=100)})

# Calculate the mean and standard deviation of the data
mean = data['value'].mean()
std = data['value'].std()

# Create the plot using Seaborn
sns.set(style="darkgrid")
ax = sns.lineplot(x="time", y="value", data=data)
ax.axhline(mean, color='red', linestyle='--', label='Mean')
ax.axhline(mean + 4 * std, color='green', linestyle='--', label='Mean + 4 Std Dev')
ax.axhline(mean - 4 * std, color='purple', linestyle='--', label='Mean - 4 Std Dev')
ax.legend(loc='best')

# Rotate the x-axis labels by 45 degrees
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right');  
# the above last ';' removes unwanted terminal output 
```
